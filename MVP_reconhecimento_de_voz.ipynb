# 6. Atibutos do dataset

O primeiro dataset a ser explorado apresenta 3168 ocorrências, divididas igualmente entre as classes "masculino" e "feminino". Cada registro representa uma gravação de voz pré-processada com o auxílio do programa R, levando em consideração o alcance da voz humana: de 0hz a 280hz. Esta coletânea detalha também 20 atributos numéricos de ordem acústica e 1 atributo categórico. Não há valores nulos em nenhuma das colunas. A seguir, descrevem-se tais atributos:


*   Meanfreq: frequência média do espectro da voz (em kHz). Importante destacar que as medidas apresentadas em cada linha das colunas deste dataset foram obtidas a partir da análise de um único espectro de voz. Isto é, a partir daquele sinal de voz obteve-se a frequência média observando a intensidade de uma determinada frequência. Também convém ressaltar que a voz masculina alcança as faixas de frequência de 85 a 180 Hz (frequências mais graves), enquanto a voz feminina	alcança valores de 165 a 255 Hz (frequências mais agudas);

*   SD: desvio padrão da frequência;

*   Median: valores de mediana das frequências (em kHz);

*   Q25: primeiro quartil da frequência do arquivo de voz (em kHz);

*   Q75: terceiro quartil da frequência (em kHz);

*   QR: amplitude interquartil (em kHz);

*   Skew: indica a assimetria da distribuição da frequência. Auxilia a descrever qualitativamente o sinal de vibração (adimensional);

*   Kurt: do inglês, *kurtosis*, pode ser traduzida como a distorção do sinal, quantificando os picos ou achatamentos deste. Também auxilia a descrever qualitativamente o sinal de vibração (adimensional);


*   Sp.ent: entropia espectral, isto é, a medida da desordem das frequências no espectro de voz (adimensional);

*   Sfm: do inglês, *spectral flatness* é a medida que auxilia a quantificar se o espectro é plano (um ruído branco) ou com picos (alterações de frequência). Esta é uma variável adimensional;

*   Mode: frequência modal. Não é a estatística clássica, mas sim a frequência mais predominante no espectro durante o registro (em kHz);

*   Centroid: indica em que ponto está o "centro de massa" do espectro (em kHz);

*   Peakf: indica os valores do pico das frequências, isto é, as frequências com mais energia (em kHz);

*   Meanfun: média da frequência fundamental medida através do sinal acústico (em kHz);

*   Minfun: frequência fundamental mínima medida através do sinal acústico (em kHz);

*   Maxfun: frequência fundamental máxima medida em um sinal acústico (em kHz);

*   Meandom: média da frequência dominante medida em um sinal acústico (em kHz);

*   Mindom: mínimo da frequência dominante medido em um sinal acústico (em kHz);

*   Maxdom: máximo da frequência dominante medido em um sinal acústico (em kHz);

*   Dfrange: amplitude da frequência dominante medida em um sinal acústico (em kHz);

*   Modindx:  índice de modulação. Calculado como a diferença absoluta acumulada entre medições adjacentes de frequências fundamentais dividida pela amplitude de frequência (adimensional);

*   Label: apresenta as classes binárias "male" ou "female" para a classificação das amostras.

Por fim, cabe pontuar a diferença de unidades de medida das variáveis. KHz significa Kilohertz e é usada para os atributos relacionados às medidas de frequência. Quanto às demais sinalizadas como "adimensionais", pode-se dizer que não apresentam unidades de medida, pois são obtidas a partir de cálculos matemáticos sobre a distribuição do sinal sonoro.

## 8.2. Valores duplicados

Nesta secção iremos explorar a possível existência de valores duplicados no dataset.

```python
#Verificação de linhas duplicadas
duplicadas = df[df.duplicated(keep=False)]

#Organizando a visualização
print(f"Número de linhas duplicadas: {len(duplicadas)}")
if not duplicadas.empty:
    print("\nLinhas duplicadas:")
    display(duplicadas.sort_values(by=df.columns.tolist()))
```

A partir da verificação feita acima, percebemos uma entrada duplicada para o gênero masculino e outra para o gênero feminino. A seguir, verificaremos a possibilidade de colunas duplicadas.

```python
# Listar pares de colunas com valores idênticos
colunas_duplicadas = []
for i in range(len(df.columns)):
    for j in range(i + 1, len(df.columns)):
        col1, col2 = df.columns[i], df.columns[j]
        if df[col1].equals(df[col2]):
            colunas_duplicadas.append((col1, col2))

if colunas_duplicadas:
    print("Colunas com valores idênticos:")
    for par in colunas_duplicadas:
        print(f"  - {par[0]} e {par[1]}")
else:
    print("Nenhuma coluna duplicada encontrada.")
```

Foram encontradas duas colunas com valores idênticos: meanfreq e centroid. A seguir, será investigado mais a fundo para perceber esta possível igualdade de valores:

```python
#Verificar possível igualdade numérica
sao_iguais = (df['meanfreq'] == df['centroid']).all()
print(f"As colunas são idênticas? {sao_iguais}")

#Cálculo da diferença média entre os valores
diferenca_media = (df['meanfreq'] - df['centroid']).mean()
print(f"Diferença média: {diferenca_media:.10f}")
```

```python
#Gráfico para observar a relação entre as variáveis
print(df[['meanfreq', 'centroid']].describe())
sns.scatterplot(data=df, x='meanfreq', y='centroid', hue='perceived_gender')
plt.title("Relação entre meanfreq e centroid")
plt.show()
```

Por meio desta breve análise exploratória, pode-se perceber a igualdade entre os valores das duas colunas. Em teoria, tais medidas deveriam ser diferentes: cálculos de meanfreq baseiam-se na média aritmética das frequências, enquanto para centroid seria usada a média ponderada ao longo da energia espectral. Provavelmente, houve algum falha ao processar os dados. Como não há informação sobre o assunto na página Kaggle do dataset e provou-se a total semelhança das informações, optou-se por excluir a coluna centroid.


```python
#Removendo a coluna 'centroid' (redundante com 'meanfreq')
df = df.drop(columns=['centroid'])

#Removendo as linhas duplicadas
linhas_antes = len(df)
df = df.drop_duplicates()
linhas_removidas = linhas_antes - len(df)

#Resultado final
print(f"Coluna 'centroid' removida com sucesso.")
print(f"{linhas_removidas} linhas duplicadas foram removidas.")
print(f"\nDimensões finais do dataset: {df.shape} (linhas, colunas)")
```

## 9.1. Média

A primeira estatística apresentada é a média dos valores dos atributos do dataset. Destaca-se que esta estatística será apresentada levando em consideração a divisão por gêneros.

```python
#Separar por gênero e calcular média
estatisticas = df.groupby("perceived_gender").agg(['mean'])

#Organizar melhor os nomes das colunas (multi-index flatten)
estatisticas.columns = [f'{col[0]}_{col[1]}' for col in estatisticas.columns]

#Separar as tabelas por gênero
masculino = estatisticas.filter(like='male', axis=0)
feminino = estatisticas.filter(like='female', axis=0)

#Transpor e reorganizar em uma única tabela
tabela_final = pd.DataFrame()
tabela_final["Média - Masc"] = df[df["perceived_gender"] == "male"].mean(numeric_only=True)
tabela_final["Média - Fem"] = df[df["perceived_gender"] == "female"].mean(numeric_only=True)

#Exibir tabela
print("Tabela comparativa por gênero:")
print(tabela_final)
```

Os valores de média obtidos são em sua maioria abaixo de 10. No entanto, é muito importante relembrar, conforme exposto na secção 6 deste MVP, que as unidades de medidas são diferentes para as as variáveis analisadas. *Skew*, *kurt*, *sp.ent*, *sfm* e *modindx* são adimensionais. Ao passo que as demais estão em KHz.

Além disso, é importante fazer um pequeno comentário acerca da variável *kurt*, isto é, a curtose. Neste MVP, trata- se de uma medida espectral da voz, e não uma medida estatística. Por este motivo, os valores altos das médias de curtose não devem ser analisados da forma tradicional.

A seguir, são apresentados os gráficos das médias obtidas para cada variável. Optou-se por expô-los separadamente com o objetivo de evitar confusões relativamente às unidades de medida.


```python
#Lista de variáveis
variaveis = tabela_final.index.tolist()

#Dados
valores_masc = tabela_final["Média - Masc"]
valores_fem = tabela_final["Média - Fem"]

#Layout do grid
n_cols = 3
n_rows = -(-len(variaveis) // n_cols)  # Arredondamento pra cima

#Criar subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 3))
axes = axes.flatten()  # Facilita iteração

#Plotar cada gráfico individualmente
for i, var in enumerate(variaveis):
    ax = axes[i]
    ax.bar(['Masculino', 'Feminino'], [valores_masc[var], valores_fem[var]], color=['steelblue', 'coral'])
    ax.set_title(var)
    ax.set_ylim(bottom=0)  # Começa do zero para facilitar comparação

#Ocultar subplots extras, se houver
for j in range(len(variaveis), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.suptitle('Média das Variáveis por Gênero', fontsize=16, y=1.02)
plt.show()
```

De maneira geral, podemos percerber que as médias das variáveis relativas às frequências -  *meanfreq, meanfun, minfun, maxfun, maxdom e dfrange* - são mais altas para o gênero feminino. Tal observação apenas reafirma a ideia de que as vozes femininas são mais altas. Isto se confirma com os valores igualmente mais altos para tal gênero nas variáveis *median, Q25, Q75 e mode*.

Quanto aos resultados de *sd*, *IQR* e *sfm*, nota-se valores maiores para as vozes masculinas. Isto poderia mostrar uma maior variação espectral para este gênero.

Na mesma linha, os valores *skew* e *kurt* também são maiores para as vozes masculinas. Com este resultado, pode-se extrair duas conclusões: ou há mais picos ou estamos diante de outliers.

Para *sp. ent* e *sfm*, o gênero masculino também apresenta os valores mais elevados. Com isso, compreende-se que este tipo de voz tem uma distribuição de frequência menos uniforme, podendo estar associada a uma ressonância vocal mais baixa. Além disso, o espectro da voz seria mais achatado, podendo indicar frequências mais baixas.

Por fim, a variável *modindx* é muito semelhante entre os gêneros. Com isso, entende-se que a questão da modularidade da voz pode não ser tão relevante para a identificação dos gêneros.

Observando os gráficos acima, percebe-se um grande número de outliers para algumas variáveis, como *kurt*. Com o código a seguir, identificar-se-á a quantidade precisa de outliers por atributo:

```python
#Contando outliers por variável
def count_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers)

#Aplicando a todas as colunas numéricas
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
outliers_per_var = {col: count_outliers(df, col) for col in numeric_cols}

print("Outliers por variável:")
for var, count in outliers_per_var.items():
    print(f"{var}: {count} outliers")
```

Para as variáveis com 0 outliers, nenhum tratamento será aplicado. Para aquelas com até 65 outliers, considera-se uma variação própria das vozes humanas. Já as variáveis *kurt* e *modindx* serão excluídas do dataset por apresentarem muitos outliers e por já terem demonstrado um comportamento pouco preditivo. O mesmo processo será aplicado a *maxfun* e *mindom*:

## 10.1. Padronização dos dados

Nesta secção será feita a padronização dos dados. Esta ação torna-se importante, pois, nas secções anteriores, constatou-se que a variável *skew* têm valores maiores, enquanto outras variáveis, como *minfun*, dados numéricos menores. Esta discrepância poderia ser lida por um algoritmo de Machine Learning de maneira errônea, atribuindo importância maior aos atributos de maior escala. Para evitar que tal situação ocorra, entra em cena a padronização.

```python
#Padronizar com StandardScaler
scaler = StandardScaler()

#Ajustes no treino e transformação treino e teste
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#Converter os arrays de volta para DataFrames (opcional, mas ajuda a visualizar)
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)

#Adicionar a variável de gênero
X_train_scaled_df['perceived_gender'] = y_train.reset_index(drop=True)
X_test_scaled_df['perceived_gender'] = y_test.reset_index(drop=True)

#Estatísticas por gênero (média e desvio padrão)
print("Estatísticas por gênero no conjunto padronizado (X_train_scaled):")
estatisticas_genero = X_train_scaled_df.groupby('perceived_gender').agg(['mean', 'std'])
print(estatisticas_genero.T)
```

Após a aplicação do StandardScaler, obtivemos valores de desvio padrão próximos a 1 e médias próximas a 0, em todo o conjunto de dados. No entanto, olhando para as estatísticas dentro de cada grupo separado por gênero estes valores são diferentes. Na verdade, ao fazer essa análise, busca-se compreender qual seria a variabilidade dos valores padronizados dentro de cada subgrupo.

Com base nos valores obtidos, pode-se também fazer algumas análises. Observa-se que para todas as variáveis, as classes masculino e feminino têm valores simetricamente opostos. Na prática, isto ocorre por conta do equilíbrio do dataset. Seria um problema apenas se fosse verificado *data leakage*, o que não ocorreu.

Verifique a seguir um exemplo com o auxílio da representação boxplot:

```python
#Observando exemplo com a variável #meanfun"
# Definir paleta colorblind-friendly (feminino: amarelo/laranja, masculino: azul)
palette = {"female": "#FFA500", "male": "#0072B2"}  # Laranja (#FFA500) e Azul (#0072B2)

#Criar o boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(
    data=X_train_scaled_df,
    x="perceived_gender",
    y="meanfun",
    palette=palette,
    order=["female", "male"]
)

#Adicionar títulos e rótulos
plt.title("Distribuição de 'meanfun' por Gênero (Dados Padronizados)", fontsize=14)
plt.xlabel("Gênero", fontsize=12)
plt.ylabel("meanfun (padronizado)", fontsize=12)
plt.xticks(ticks=[0, 1], labels=["Feminino", "Masculino"])

#Remover bordas para um visual mais limpo
sns.despine()

plt.show()
```
